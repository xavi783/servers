```python
import numpy as np
import pandas as pd
import datetime as dt

import requests
from StringIO import StringIO as reader

import multiprocessing as mp
import pandas.io.data as web
```

Taht seems to be a simple question but indeed it was a complicated problem to solve because I didn't know about any web or data provider which give you this information in a easy way, so after a while searching for  this data I found a web that shows a link with this data, I speak about NASDAQ officcial web page that you can se [here](http://www.nasdaq.com/screening/companies-by-industry.aspx?&exchange=NASDAQ).

We will take the CSV files fro NYSE market from [here](http://www.nasdaq.com/screening/companies-by-industry.aspx?exchange=NYSE&render=download) but we will do it programatically with **requests** module and process with pandas at same time

### Getting Sectors

```python
_path = 'http://www.nasdaq.com/screening/companies-by-industry.aspx?exchange=NYSE&render=download'
raw_data = pd.read_csv(reader(requests.get(_path).text))
raw_data.head(1).append(raw_data.tail(1))
```

Grouping by sectors:

```python
groups = raw_data.groupby('Sector').count().iloc[:-1,:-1]
groups
```

Hereby, we will get the close of each symbol in each secotr which we want to analyze (all sectors excep n/a)

### Retrieving prices

Once we have sectors, we will get the prices from yahoo with pandas package, but we have a big amount of data to deal so we will open a HDF5 storing and we will try to save data inside it.

First step consist on open the HDF5 store, we notice that HDF5 format allows concurrent work, it means that we can read and write from different parallel process, so we will use multiprocessing package to speed up our calculations.

```python
store = pd.HDFStore('./store.h5')
store.close()
```

If you want to know more about HDF5 format, you can check the following links:

* [Introduction to HDF5](http://lopezrj.github.io/hdf5/programming/2014/09/06/hdf5.html)
* [PyTable - Python Package](http://www.pytables.org/moin/HowToUse)
* [HDF5 Group](http://www.hdfgroup.org/HDF5/)
* [Wikipedia](http://en.wikipedia.org/wiki/Hierarchical_Data_Format)

You have to take into account that the object `pd.HDFStore` only has a subset of methods of PyTables, so you don't have all the power of PyTables but it is enough for our purposes here because we only want to save Panda's DataFrames in an easy way

```python
start_date = (2000,1,1)
sector = 'Miscellaneous'
_close = []
_error = []

_symbols = raw_data['Symbol'][raw_data['Sector']==sector].values
for i, symbol in enumerate(_symbols):
    try:
        symbol = symbol.replace(' ','')
        ds = web.get_data_yahoo(symbol, dt.datetime(*start_date))['Close']
        ds.to_hdf('./store.h5','/'+sector+'/'+symbol)
    except IOError as err:
        _error.append('#'+str(i)+': '+err.message)
_error
```
